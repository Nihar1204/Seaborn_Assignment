## Seaborn Assignment

```python
# Q1

# (1) Line Plot:
# Use: Display trends over a continuous interval or time series data.

# (2) Bar Plot:
# Use: Compare categories of data, especially when one axis represents categorical data and the other represents 
# numerical values.

# (3) Histogram:
# Use: Show the distribution of a single numeric variable. Helpful in understanding data distribution and identifying
# outliers.

# (4) Box Plot (or Box-and-Whisker Plot):
# Use: Visualize the distribution of a continuous variable through displaying its summary statistics 
# (median, quartiles, outliers).

# (5) Violin Plot:
# Use: Combine aspects of box plot and kernel density estimation to show the distribution of a numeric variable for
# different categories.
```

```python
# Q2

import seaborn as sns
import matplotlib.pyplot as plt


fmri_data = sns.load_dataset("fmri")

plt.figure(figsize=(10, 6))  

sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region", markers=True)


plt.title('FMRI Signal over Time for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.tight_layout()
plt.show()
```

```python
# Q3

titanic_data = sns.load_dataset("titanic")

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Box plot for 'pclass' vs 'age'
sns.boxplot(x='pclass', y='age', data=titanic_data, ax=axes[0])
axes[0].set_title('Passenger Class vs Age')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Age')

# Box plot for 'pclass' vs 'fare'
sns.boxplot(x='pclass', y='fare', data=titanic_data, ax=axes[1])
axes[1].set_title('Passenger Class vs Fare')
axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Fare')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
# Q4

diamonds_data = sns.load_dataset("diamonds")

plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', bins=30, multiple='stack')
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title='Cut')
plt.show()
```

```python
# Q5

# Load the iris dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with 'species' as hue
sns.pairplot(iris_data, hue='species')
plt.title('Pair Plot of Iris Dataset')
plt.show()
```

```python

flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix for the heatmap
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Passenger Counts by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
